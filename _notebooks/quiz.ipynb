```python
# import useful commands
import sys
import getpass

# Counter for end score
correctanswer = 0
wronganswer = 0
totalanswer = 0

# each question is defined as a separate function
# question 1
def question1():
    global correctanswer
    global wronganswer
    global totalanswer
    # global ___ is supposed to make the thing work and not give the local variable referenced before assignment error but it aint workin
    answertoq1 = input("Question 1: What is Python's output command?")
    if answertoq1.lower() == "print":
        print("You answered " + answertoq1 + ". Well done.")
        correctanswer += 1
        totalanswer += 1
        return correctanswer and totalanswer
    else:
        print("You answered " + answertoq1 + ". Incorrect.")
        wronganswer += 1
        totalanswer += 1
        return wronganswer and totalanswer

# question 2
def question2():
    global correctanswer
    global wronganswer
    global totalanswer
    answertoq2 = input("What is the Python command to define a function?")
    if answertoq2.lower() == "def":
        print("You answered " + answertoq2 + ". Well done.")
        correctanswer += 1
        totalanswer += 1
        return correctanswer and totalanswer
    else:
        print("You answered " + answertoq2 + ". Incorrect.")
        wronganswer += 1
        totalanswer += 1
        return wronganswer and totalanswer

# question 3
def question3():
    global correctanswer
    global wronganswer
    global totalanswer
    answertoq3 = input("What is the Python command to get a user input?")
    if answertoq3.lower() == "input":
        print("You answered " + answertoq3 + ". Well done.")
        correctanswer += 1
        totalanswer += 1
        return correctanswer and totalanswer
    else:
        print("You answered " + answertoq3 + ". Incorrect.")
        wronganswer += 1
        totalanswer += 1
        return wronganswer and totalanswer

# question 4
def question4():
    global correctanswer
    global wronganswer
    global totalanswer
    answertoq4 = input("What is the parameter that a function takes called?")
    if answertoq4.lower() == "prompt" or answertoq4.lower() == "argument":
        print("You answered " + answertoq4 + ". Well done.")
        correctanswer += 1
        totalanswer += 1
        return correctanswer and totalanswer
    else:
        print("You answered " + answertoq4 + ". Incorrect.")
        wronganswer += 1
        totalanswer += 1
        return wronganswer and totalanswer

# question 5

    
# the actual quiz
# asks whether or not user wants to take quiz, gives questions if yes, ends if no
yesno = input("Welcome, " + str(getpass.getuser()) + ", to this quiz. Would you like to attempt?")
if yesno.lower() == "yes":
    print("Good. Proceed.")
    question1()
    question2()
    question3()
    question4()
else:
    print("Goodbye.")
    # sys.exit()
    # ^ is not working, try to find another way to exit system. arthur suggested "except SystemExist:" but didnt work but maybe try more
    # but its not really needed tbf, can just end at print

# make grade score calculator here, move this and the "test the counters" to under the if yes part after the questions
percentcorrect = 100 * correctanswer / totalanswer
percentwrong = 100 * wronganswer / totalanswer
if percentcorrect >= 90:
    print("")

# test the counters
print("number of correct answers is " + str(correctanswer))
print("number of wrong answers is " + str(wronganswer))
print("number of total answers is " + str(totalanswer))
```
